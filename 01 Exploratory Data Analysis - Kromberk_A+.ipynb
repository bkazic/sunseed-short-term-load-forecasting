### Normalizing data

```python
from sklearn import preprocessing

x = load_data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data_norm = pd.DataFrame(x_scaled, index=load_data.index, columns=load_data.columns)
#data_norm.head()
ax = data_norm.plot(figsize=(20,6))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
data = load_data['2011-01-9':'2011-02-6'][["000002637", "000129728","000137187"]]

#plot = data.plot(subplots=True, layout=(3,1), style=['b','b','b'], figsize=(20,14), x_compat=True)

fig, axx = plt.subplots(data.shape[1], figsize=(20,12))
for x, columns in enumerate(data):
    axx[x].plot_date(data.index.to_pydatetime(), data[columns], fmt='-', label=columns)

# Set major x ticks on Mondays.
for ax in axx:  
    #ax=plt.gca()
    ax.xaxis.set_major_locator(dates.HourLocator(byhour=12))
    ax.xaxis.set_major_formatter(dates.DateFormatter('%a'))
    ax.legend()
    ax.grid()

plt.tight_layout()
```

By looking at the data of 3 randomly seleceted weeks from January, we can observe some pattern. We can see that the weekends are indeed different from the week. But we can still see that data from week to week can be as well very differetn. What is the cause? Weather?

## Daily Trend

```python
dates.DayLocator()
```

```python
week_data = pd.DataFrame.pivot_table(load_data, index=load_data.index.hour, columns= load_data.index.dayofweek, values=["000002637"])
week_data.head()

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

ax = week_data.plot(figsize=(8,5))
ax.legend(labels=days)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
datasets = {
    "Summer data": load_data['2014-05-1':'2014-09-1'],
    "Winter data": load_data['2014-10-1':'2015-02-18']
}

f, axx = plt.subplots(1, len(datasets), sharey=True, figsize=(14,5))
for idx, (key, dataset) in enumerate(datasets.iteritems()):

    data = pd.DataFrame.pivot_table(dataset, index=dataset.index.hour, columns= dataset.index.dayofweek, values=["000002637"])
    ax = data.plot(ax = axx[idx], title=key)
    
    ax.legend(labels=days)
    ax.set_xlabel("Hours")
        
axx[0].set_ylabel("Load [kW]")
plt.tight_layout()
```

## Weekely Trend

```python
import seaborn as sns 

plt.figure(figsize=(8,5))

sns.set_style("whitegrid")
ax = sns.violinplot(week_data)
ax.set_xticklabels(days, rotation='vertical')
```

## Yearly Trend

```python
month_data = pd.DataFrame.pivot_table(load_data, index=load_data.index.month, columns= load_data.index.dayofweek, values=["000002637"])
months = ['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.heatmap(month_data["000002637"], linewidths=0, cmap="RdBu_r", yticklabels=months, xticklabels=days) # Plot temperature heatmap
_ = plt.xticks(rotation=90) 
```
