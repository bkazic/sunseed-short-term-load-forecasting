## Data Auditing

**Explenation of meterological variables** (ref: http://meteo.arso.gov.si/uploads/meteo/help/en/razlaga_spremenljivk.html)
- **p** - mean pressure (hPa): Daily mean of air pressure(pmean) is calculated as arithmetic mean of measured air pressure at 7 a.m., 2 p.m. and 9 p.m. (CET): pmean=(p7+p14+p21)/3 Monthly and annual mean of air pressure is arithmetic mean of the daily means.
- **pmin** - minimum pressure(hPa): The lowest pressure within the given time period.
- **pmax** - maximum pressure(hPa): The highest pressure within the given time period.
- **t2m_termin** - temperature at 2m (°C): Air temperature 2m above the ground.
- **t2m** - mean temperature at 2m (°C): Mean air temperature (Tmean) 2m above the ground is calculated as weighting mean of measured temperatures at 7 a.m., 2 p.m. and 9 p.m. (CET) 2 m above the ground (inside the white and wooden Stevenson screen): Tmean=(T7+T14+2T21)/4 (Kämtz, 1860) Monthly and annual mean of temperature is arithmetic mean of the daily means.
- **t2mmin** - minimum air temperature 2 m above the ground (°C): The lowest temperature 2 m above the ground within the given time period.
- **t2mmax** - maximum air temperature 2 m above the ground (°C): The highest temperature 2 m above the ground within the given time period.
- **rh_termin** - relative humidity (%): Relative humididty.
- **rh** - mean relative humidity (%): Mean relative humidity (rhmean) is calculated as arithmetic mean of measured relative humidity at 7 a.m., 2 p.m. and 9 p.m. (CET): rhmean=(rh7+rh14+rh21)/3 Monthly and annual mean of relative humidity is arithmetic mean of the daily means.
- **rhmin** - minimal relative humidity (%): The lowest humidity within the given time period.
- **rhmax** - maximum relative humidity (%): The highest humidity within the given time period.
- **padavine** - precipitation accumulation (mm): Accumulated precipitation within the given time period.
- **veter_hitrost** - mean wind speed (m/s): Daily average of wind speed (vmean) is calculated as arithmetic mean of measured wind speed at 7 a.m., 2 p.m. and 9 p.m. (CET): vmean=(v7+v14+v21)/3 Monthly and annual mean of wind speed is arithmetic mean of the daily means.
- **veter_vek_smer** - mean wind direction (°): Mean wind direction.
- **veter_max_hitrost** - max wind gust (m/s): Max wind gust.
- **energija_gl** - mean global energy radiation (W/m2): Mean global energy radiation.
- **energija_di** - mean diffusive energy radiation (W/m2): Mean diffusive energy radiation.

According to literature review survey (Guerci et al., 2010), it is a well-known that meteorological conditions have a significant influence on electricity loads. The main weather temperature that is usually used in studies and is probably the most important is temperature. Some literature report the the use of other variables such as solar radiation, humidity, wind speed, cloudiness, rainfall.

```python
import seaborn as sns 

temperature = pd.DataFrame.pivot_table(weather_data, index=weather_data.index.month, columns= weather_data.index.day, values='t2m')
months = ['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
_ = sns.heatmap(temperature, linewidths=0, cmap="RdBu_r", yticklabels=months) # Plot temperature heatmap
```

```python
import seaborn as sns 

temperature = pd.DataFrame.pivot_table(weather_data, index=weather_data.index.month, columns= weather_data.index.dayofweek, values='t2m')
months = ['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
_ = sns.heatmap(temperature, linewidths=0, cmap="RdBu", yticklabels=months, xticklabels=days) # Plot temperature heatmap
_ = mpl.pyplot.xticks(rotation=90) 
_ = mpl.pyplot.title("Temperature heatmap")
```

If we do not use reversed heatmap (cmap="RdBu"), than we can see similar heatmap as we got for electricity load. This shows that the anual seasonality in electricity load are mainly governed by climate conditions, such as outdor temperature.
