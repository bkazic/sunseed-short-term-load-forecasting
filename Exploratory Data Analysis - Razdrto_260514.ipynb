```python
%matplotlib inline
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
```

Import data...

```python
load_data = pd.read_csv(r".\data\Sunseed_razdrto260514.csv", index_col=0, parse_dates=True)
load_data.head()
```

```python
load_data.describe()
```

NOTE: Datetime index is not ordered!!

```python
data = load_data['2013-01-9':'2013-02-6'][["89369517", "89369524","89369597"]]
data = data.sort_index()
#plot = data.plot(subplots=True, layout=(3,1), style=['b','b','b'], figsize=(20,14), x_compat=True)

fig, axx = plt.subplots(data.shape[1], figsize=(20,12))
for x, columns in enumerate(data):
    axx[x].plot_date(data.index.to_pydatetime(), data[columns], fmt='-', label=columns)

# Set major x ticks on Mondays.
for ax in axx:  
    #ax=plt.gca()
    ax.xaxis.set_major_locator(dates.HourLocator(byhour=12))
    ax.xaxis.set_major_formatter(dates.DateFormatter('%a'))
    ax.legend()
    ax.grid()

plt.tight_layout()
```
