```python
#input_file_name = "mergedLinearly.csv"
input_file_name = "mergedLinearlySorted.csv"
input_file_path = os.path.join(os.path.join(os.getcwd(), "data"), input_file_name)

data = pd.read_csv(input_file_path, index_col=0, parse_dates=True)
originalSize = data.shape[1]
data.head()
```

## Shifting Target Data

Shifting the target columns, according to prediction horizon.

```python
prediction_horizon = -10

#columnsToShift = data.iloc[:,0:-7].columns.tolist()
columnsToShift = list(data)
shiftedColumns = [str(prediction_horizon)[1:] + "h_" + col for col in columnsToShift]

data[shiftedColumns] = data[columnsToShift].shift(prediction_horizon, freq='H')
data.dropna(inplace=True)
data.head()
data.shape
```

```python
import seaborn as sns
sns.set(context="paper", font="monospace")

corrmat = data.corr()
#corrmat = data['2013-01-07 22:00:00':'2014-01-14 22:00:00'].corr()
#corrmat = data['2014-01-07 22:00:00':'2014-01-14 22:00:00'].corr()

f, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(corrmat, vmax=.8, linewidths=0, cmap="coolwarm", square=True)
#sns.heatmap(corrmat, vmax=.8, linewidths=0, square=True)

# draw white lines between datasets
size = corrmat.shape[1]
#ax.axhline(size-16, c="w")
ax.axvline(16, c="w")
#ax.axhline(size-30, c="w")
ax.axvline(30, c="w")
#ax.axhline(size-47, c="w")
ax.axvline(47, c="w")
ax.axvline(originalSize, c="w")
#ax.axhline(size + size-16, c="w")
ax.axvline(originalSize + 16, c="w")
#ax.axhline(size + size-30, c="w")
ax.axvline(originalSize + 30, c="w")
#ax.axhline(size + size-47, c="w")
ax.axvline(originalSize+ 47, c="w")

plt.title("Correlations between features (from different data sets)")
#f.tight_layout()

plt.savefig("./figures/correlations_{}h.pdf".format(str(prediction_horizon)[1:]))
plt.savefig("./figures/correlations_{}h.png".format(str(prediction_horizon)[1:]), dpi=300)
```
