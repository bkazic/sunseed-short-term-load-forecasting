```python
%matplotlib inline
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import numpy as np
import os
```

### Loading files

Importing electricity data.

```python
#input_file_name = "Kromberk_A+.csv"
#input_file_name = "estimatedNodeMeasurements.csv"
input_file_name = "estimatedNodeMeasurementsSorted.csv"
input_file_path = os.path.join(os.path.join(os.getcwd(), "data"), input_file_name)

load_data = pd.read_csv(input_file_path, index_col=0, parse_dates=True)
load_data.head()
#load_data.shape
```

Importing weather data.

```python
orderedIndex = list(load_data.columns.values)
subOrderedIndex = list(load_data.iloc[:,14:30].sum().order(ascending=False).index)
orderedIndex[14:30] = subOrderedIndex
orderedIndex = orderedIndex[-16:] + orderedIndex[:-16] # Move end nodes to the begining
load_data = load_data.reindex(columns=orderedIndex)
load_data.head()
```

```python
input_file_name = "Bilje.csv"
input_file_path = os.path.join(os.path.join(os.getcwd(), "data"), input_file_name)

weather_data = pd.read_csv(input_file_path, index_col=0, parse_dates=True)
weather_data.head()
weather_data.shape
```

### Merging

```python
#merged = pd.concat([load_data, weather_data], axis=1) # not working
#merged = pd.concat([load_data, weather_data], axis=1, join_axes=[load_data.index]) # not working
#merged = pd.concat([load_data, weather_data], axis=1, join_axes=[load_data.index], join='inner') # not working

#merged = load_data.join(weather_data)
merged = load_data[:'2015-02-01'].join(weather_data)

merged.head()
```

### Interoplation

```python
interpolated = merged.interpolate()
#interpolated = merged.interpolate(method='spline', order=2) # takes a long time
interpolated = np.round(interpolated, decimals=2)
interpolated.dropna(inplace=True)

interpolated.head()
```

## Exporting to csv

```python
#output_file_name = "mergedLinearly.csv"
#output_file_name = "mergedLinearlySorted.csv"
#output_file_name = "mergedSpline.csv"
output_file_name = "merged.csv"
output_file_name = os.path.join(os.path.join(os.getcwd(), "data"), output_file_name)

interpolated.to_csv(output_file_name)
```
